```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('CRSP')
df
```

```python
df = df.set_index(['tic','rdq'])
df.columns = df.columns.str.upper()
df
```

```python
(~(df['TEQQ'].isna())).sum()
```

```python
columns = ['FYEARQ', 'FQTR', 'CONM'] + list(set(['IVNCFY', 'FINCFY', 'OANCFY', 'CHQ', 'PRCCQ',
                    'DLCQ', 'DLTTQ', 'TEQQ', 'SALEQ', 'MKVALTQ', 'EPSPI12', 'CSH12Q',
                    'DVPSPQ', 'OANCFY', 'CAPXY', 'ATQ', 'LTQ', 'CHEQ', 'RSTCHELTQ', 'ACTQ',
                    'PPENTQ', 'CSHIQ', 'IVAOQ', 'DLCQ', 'DLTTQ', 'MIBQ', 'PSTKQ', 'CEQQ',
                    'PPEGTQ', 'INVTQ', 'AJEXQ', 'CSHOQ', 'TXDITCQ', 'TEQQ', 'PSTKQ', 'CSHOQ', 'NIQ', 'DPQ',
                    'WCAPCHY', 'IBQ', 'TXDBQ', 'SEQQ']))
```

```python
df_new = df[columns]
import gc
gc.collect()
df_new
```

Need to group by tic

```python
df_new = df_new.sort_index()
df_new = df_new.loc[df_new.index.dropna()]
df_new = df_new.drop_duplicates()
```

```python
df_new['NOA'] = (df_new['ATQ']-df_new['CHEQ'].fillna(0)-df_new['IVAOQ'].fillna(0)) \
    -(df_new['LTQ']-df_new['DLCQ'].fillna(0)-df_new['DLTTQ'].fillna(0)- \
      df_new['MIBQ'].fillna(0)-df_new['PSTKQ'].fillna(0)-df_new['CEQQ'].fillna(0))

df_new['BEQ'] = df_new['TEQQ'] + df_new['TXDITCQ'].fillna(0) - df_new['PSTKQ'].fillna(0)
```

```python
df2 = []

features = ['CF2P', 'LEV', 'S2P', 'Q', 'E2P', 'D2P', 'CF', 'C', 'BEME', 'A2ME', 'INVESTMENT', 'NOA',
           'DPI2A', 'NI']

for tic, sub in df_new.groupby('tic'): # .loc[['AAPL','GME']]
    sub = sub.copy()
    gc.collect()

    sub['CF2P'] = (sub['IBQ'].rolling(4).sum(skipna=False) + \
                   sub['DPQ'] + 
                   sub['TXDBQ'])/ (sub['CSHOQ']*sub['PRCCQ'])
    
    sub['LEV'] = (sub['DLCQ'] + sub['DLTTQ']) / (sub['DLCQ'] + sub['DLTTQ'] + sub['SEQQ'])
    
    sub['S2P'] = (sub['SALEQ'].rolling(4).sum(skipna=False)) / (sub['CSHOQ']*sub['PRCCQ'])
    
    sub['Q'] = (sub['ATQ'] + (sub['CSHOQ']*sub['PRCCQ']) - sub['CHEQ'] - sub['TXDBQ'])/sub['ATQ']
    
    sub['E2P'] = sub['IBQ'].rolling(4).sum(skipna=False) / (sub['CSHOQ']*sub['PRCCQ'])
    
    sub['D2P'] = (sub['DVPSPQ'].rolling(4).sum(skipna=False))/sub['PRCCQ']
    
    sub['BEME'] = sub['BEQ'] / (sub['CSHOQ']*sub['PRCCQ'])
    
    sub['CF'] = (sub['NIQ'] + sub['DPQ'] - sub['WCAPCHY'] - sub['CAPXY']) / sub['BEME']
    
    sub['C'] = sub['CHEQ'] / sub['ATQ']
            
    sub['A2ME'] = sub['ATQ'] / (sub['CSHOQ']*sub['PRCCQ'])
    
    sub['INVESTMENT'] = (sub['ATQ']-sub['ATQ'].shift(4)) / sub['ATQ'].shift(4)
    
    sub['NOA'] = sub['NOA']
    
    sub['DPI2A'] = (sub['PPEGTQ'] + sub['INVTQ'])/sub['ATQ'].shift(4)
    
    sub['NI'] = np.log(sub['AJEXQ']*sub['CSHOQ']) - np.log(sub['AJEXQ'].shift(4)*sub['CSHOQ'].shift(4))
        
    sub = sub[['FYEARQ','FQTR','CONM']+features]
    
    df2.append(sub)
```

```python
df2 = pd.concat(df2,axis=0)
df2
```

```python
df2.to_csv('features_inv_value.csv')
```
